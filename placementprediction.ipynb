```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import LabelEncoder
```

```python
data=pd.read_csv('placementdata (1).csv')
```

```python
data.head()
```

```python
data.drop(columns="StudentID",axis=1,inplace=True)
```

```python
oe_extra = OrdinalEncoder(categories=[['No','Yes']]) # Create an OrdinalEncoder for 'ExtracurricularActivities'
oe_place = OrdinalEncoder(categories=[['No','Yes']]) # Create another OrdinalEncoder for 'PlacementTraining'

data["extra"] = oe_extra.fit_transform(data[['ExtracurricularActivities']])
data['place'] = oe_place.fit_transform(data[['PlacementTraining']])
```

```python
le=LabelEncoder()
data['placement']=le.fit_transform(data[['PlacementStatus']])
```

```python
data=data.drop(columns=['PlacementStatus'],axis=1)
```

```python
data.head()
```

```python
data.head()
```

```python
x=data.drop(columns="placement",axis=1)
x.head()
y=data['placement']
y.head()
```

```python
data=data.drop(columns=['ExtracurricularActivities','PlacementTraining'],axis=1)
```

```python
data.head()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2 ,stratify=y, random_state=1)
```

```python
print(x_train.shape,x_test.shape)
```

```python
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
traiing_predict=model.predict(x_train)
training_accuracy=accuracy_score(traiing_predict,y_train)
print(training_accuracy*100)
```

```python
test_predict=model.predict(x_test)
test_accuracy=accuracy_score(test_predict,y_test)
print(test_accuracy*100)
```

```python
predict=model.predict(y)
print(predict)
```
